```python
# Exercice 1 : Création et écriture d’un fichier 
# 1. Créez le fichier villes.txt en mode écriture ("w"). 
# 2. Écrivez-y les lignes suivantes : 


with open("villes.txt","w") as file:
    l1="Rabat,34.020882,-6.841650\n"
    l2="Casablanca,33.573110,-7.589843\n"
    l3="Fès,34.033333,-5.000000\n" 
    l4="Tanger,35.759465,-5.833954\n"
    l5="Agadir,30.427755,-9.598107\n"
    file.write(l1)
    file.write(l2)
    file.write(l3)
    file.write(l4)
    file.write(l5)
```

```python
# Exercice 2 : Lecture et affichage formaté 
# 1. Ouvrez le fichier villes.txt en lecture. 
# 2. Pour chaque ligne, affichez : 
# "La ville de [ville] est située à la latitude [lat] et la longitude [lon]." 
# 3. Comptez le nombre total de villes dans le fichier. 

villes_nombers = 0
with open("villes.txt","r") as villes:
    villes_as_lsit = villes.readlines()
    for vill in villes_as_lsit:
        villes_nombers +=1
        vill_list = vill.strip().split(",")
        print(f"La ville de [{vill_list[0] }] est située à la latitude [{vill_list[1]}] et la longitude {vill_list[2]}")
        
print("le nombre total de villes dans le fichier:",villes_nombers)
```

```python
# Exercice 3 : Création d’un dictionnaire géographique 
# 1. Lisez le fichier villes.txt. 
# 2. Créez un dictionnaire nommé coordonnees, 
# où la clé est le nom de la ville et la valeur est un tuple contenant (latitude, longitude). 
# 3. Affichez ce dictionnaire.
coordonnees={}

with open("villes.txt","r") as villes:
    villes_as_lsit = villes.readlines()
    for vill in villes_as_lsit:
        # print(vill.split(",")[0])
        coordonnees[vill.split(",")[0]] = (float(vill.split(",")[1]),float(vill.split(",")[2]))

coordonnees
```

```python
# Exercice 4 : Calcul d’une distance simplifiée 
# On souhaite maintenant calculer la distance approximative en degrés entre deux villes du dictionnaire. 
# 1. Écrivez une fonction : 
def distance(v1, v2): 
    return abs(v1[0] - v2[0]) + abs(v1[1] - v2[1]) 
# 2. Utilisez cette fonction pour afficher la distance entre Rabat et Casablanca. 

dist_rc = distance(coordonnees["Rabat"], coordonnees["Casablanca"])
print(f"Distance entre Rabat et Casablanca : {dist_rc} °")
```

```python
# Exercice 5 : Sauvegarde dans un nouveau fichier 
# Créez un fichier distances.txt contenant, pour chaque ville, la distance simplifiée depuis Rabat.

with open("distances.txt", "w") as f:
    for ville, coord in coordonnees.items():
        dist = distance(coordonnees["Rabat"], coord)
        f.write(f"{ville},{dist}\n")
        
```

```python
# ex6

print("Villes à plus de 2° de Rabat :")

with open("distances.txt", "r") as file:
    for ligne in file:
        ville, dist = ligne.strip().split(",")
        dist = float(dist)
        if dist > 2:
            print(f" - {ville} ({dist:.2f}°)")
```

```python
# Exercice 7 : Résumé général
# 1. Comptez combien de villes sont à moins de 2°, entre 2° et 5°, et au-delà de 5°.
# 2. Affichez un petit résumé sous forme de texte.

moins_de_2 = 0
entre_2_et_5 = 0
au_dela_de_5 = 0

with open("distances.txt", "r") as file:
    for ligne in file:
        ville, dist = ligne.strip().split(",")
        dist = float(dist)
        
        if dist < 2:
            moins_de_2 += 1
        elif dist <= 5:
            entre_2_et_5 += 1
        else:
            au_dela_de_5 += 1

# Affichage du résumé
print("=" * 50)
print("RÉSUMÉ DES DISTANCES DEPUIS RABAT")
print("=" * 50)
print(f"Villes à moins de 2° : {moins_de_2}")
print(f"Villes entre 2° et 5° : {entre_2_et_5}")
print(f"Villes au-delà de 5° : {au_dela_de_5}")
print("=" * 50)
print(f"Total : {moins_de_2 + entre_2_et_5 + au_dela_de_5} villes")
print("=" * 50)
```
