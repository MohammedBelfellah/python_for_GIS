```python
# Exercice 4 : Fusion de dictionnaires de population 

# On dispose de deux dictionnaires : 
pop1 = {"Rabat": 577827, "Casablanca": 3359818, "Fès": 1112072} 
pop2 = {"Casablanca": 500000, "Tanger": 947952, "Marrakech": 928850} 

# Fusionnez 
population_totale = {}
for i,j in pop1.items() | pop2.items():
    population_totale[i] = j

# Affichez le dictionnaire final. 
print(population_totale)

# Pour les clés communes, additionnez les valeurs. 
nv_dic = {}
for cle in pop1.keys() | pop2.keys():
    nv_dic[cle] = pop1.get(cle,0) + pop2.get(cle,0)
print(nv_dic)
```

```python
# Exercice 6 : Extraction des villes par pays

villes_info = { 
    "Rabat": {"latitude": 34.020882, "longitude": -6.841650, "pays": "Maroc"}, 
    "Casablanca": {"latitude": 33.573110, "longitude": -7.589843, "pays": "Maroc"}, 
    "Tanger": {"latitude": 35.759465, "longitude": -5.833954, "pays": "Maroc"}, 
    "Paris": {"latitude": 48.8566, "longitude": 2.3522, "pays": "France"} 
}

# Fonction pour filtrer les villes par pays
def villes_par_pays(dictionnaire, nom_pays):
    """Retourne une liste des noms de villes situées dans le pays spécifié"""
    return [ville for ville, info in dictionnaire.items() if info["pays"] == nom_pays]

# Test pour le pays "Maroc"
villes_maroc = villes_par_pays(villes_info, "Maroc")
print("Villes au Maroc:", villes_maroc)

# Test pour le pays "France"
villes_france = villes_par_pays(villes_info, "France")
print("Villes en France:", villes_france)
```
